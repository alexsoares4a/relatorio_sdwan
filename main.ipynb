# RELATÓRIO DE MEDICAO

```python
url_rel_medicao = r"planilhas\SMSU_TC 023_Levantamento_04540_TC 023SMSU2024_21032025_120844_V3.0.xlsx"
```

```python
df_sdwan = df_sdwan[['Prime_ID_Prodam_C', 'sequencia_C',
       '1_Nivel_Orgao_Signatario_Pagador_C', '2_Nivel_Secretaria_C',
       '3_Nivel_Coordenadoria_C', 'hierarquia_L1', 'ID_contratada_L1',
       'LOTE_1_L1', 'ID_contratada_L2', 'LOTE_2_L2', 'ID_contratada_L31',
       'LOTE_3_L31', 
       'LOTE_3_Link_Banda_Larga_1_contratada',
       'ID_contratada_L32', 'LOTE_3_L32', 'data_cancelamento_L1',
       'data_desativacao_L1', 'data_divulg_tec_relac_L1',
       'anterior_operacional_L1', 'data_servico_L1',
       'data_provisorio_identif_L1', 'data_aceite_L1', 'movimentacao_L1',
       'servico_L1', 'Contrato_Secretaria_Prodam_C',
       'Contrato_Operadora_Prodam_L1', 'Contrato_Operadora_Prodam_L2',
       'Contrato_Operadora_Prodam_L31', 'Contrato_Operadora_Prodam_L32',
       '4_Nivel_Entidade_Instalacao_Nome_Unidade_C',
       'data_desativacao_link_mpls_relacionado_C', 'Sigla_Unidade_nome_C',
       'Tipo_Logradouro_C', 'Endereco_Instalacao_logradouro_C',
       'Numero_Logradouro_C', 'Bairro_C', 'Cidade_C', 'CEP_C', 'status_L1',
       'acesso_sem_uso_C', 'faturavel_L1', 'gratuito_L1']]
```

```python
df_sdwan.columns
```

```python
list_campos_data = [
        'data_divulg_tec_relac_L1',
        'data_servico_L1',
        'data_aceite_L1']

for nome_do_campo in list_campos_data:
        #df_sdwan[nome_do_campo] = pd.to_datetime(df_sdwan[nome_do_campo]).dt.floor('D')
    df_sdwan[nome_do_campo] =  df_sdwan[nome_do_campo].dt.normalize()
```

```python
df_sdwan[list_campos_data].head()
```

## Total de Equipamentos

```python
df_total_equipamento = df_sdwan.groupby("LOTE_1_L1")[["LOTE_1_L1"]].count()

df_total_equipamento.rename_axis(index="Tipo", inplace=True)
df_total_equipamento.columns = ["Quantidade Medida"]
df_total_equipamento
```

```python
# Dicionário com os nomes atuais e os novos nomes
nomes_equipamentos = {
    "CPE-SD-WAN tipo 1 - Até 50 usuários e throughput de 150Mbps":"DISPONIBILIZAÇÃO DE EQUIPAMENTO CPE-SD-WAN TIPO 1 (PARA LOCAIS ATÉ 50 USUÁRIOS E THROUGHPUT DE 150 MBPS SIMULTÂNEO NOS SERVIÇOS DE SEGURANÇA)",
    "CPE-SD-WAN tipo 2 - Até 150 usuários e throughput de 300Mbps":"DISPONIBILIZAÇÃO DE EQUIPAMENTO CPE-SD-WAN TIPO 2 (PARA LOCAIS ATÉ 150 USUÁRIOS E THROUGHPUT DE 300 MBPS SIMULTÂNEO NOS SERVIÇOS DE SEGURANÇA)",
    "CPE-SD-WAN tipo 3 - Até 250 usuários e throughput de 500Mbps":"DISPONIBILIZAÇÃO DE EQUIPAMENTO CPE-SD-WAN TIPO 3 (PARA LOCAIS ATÉ 250 USUÁRIOS E THROUGHPUT DE 500 MBPS SIMULTÂNEO NOS SERVIÇOS DE SEGURANÇA)"
}
```

```python
# Aplica a renomeação
df_total_equipamento = df_total_equipamento.rename(index=nomes_equipamentos)
```

```python
df_total_equipamento
```

## Total de Link Simétrico a Faturar - VIVO

```python
total_link_simetrico = df_sdwan.groupby("LOTE_2_L2")[["LOTE_2_L2"]].count()

total_link_simetrico.rename_axis(index="Tipo", inplace=True)
total_link_simetrico.columns = ["Quantidade Medida"]
total_link_simetrico
```

```python
# Dicionário com os nomes atuais e os novos nomes
nomes_link_simetrico = {
    10: "LINK DE CONECTIVIDADE SIMÉTRICO 10MBPS (LCS) COM CPE",
    20: "LINK DE CONECTIVIDADE SIMÉTRICO 20MBPS (LCS) COM CPE",
    50: "LINK DE CONECTIVIDADE SIMÉTRICO 50MBPS (LCS) COM CPE",
    150: "LINK DE CONECTIVIDADE SIMÉTRICO 150MBPS (LCS) COM CPE"
}
```

```python
# Aplica a renomeação
total_link_simetrico = total_link_simetrico.rename(index=nomes_link_simetrico)
```

```python
total_link_simetrico
```

## Total de Link Banda Larga a Faturar - CLARO

```python
total_link_banda_larga = df_sdwan.groupby("LOTE_3_Link_Banda_Larga_1_contratada")[["LOTE_3_Link_Banda_Larga_1_contratada"]].count()

total_link_banda_larga.rename_axis(index="Tipo", inplace=True)
total_link_banda_larga.columns = ["Quantidade Medida"]
total_link_banda_larga
```

```python
# Dicionário com os nomes atuais e os novos nomes
nomes_link_banda_larga = {
    "LIB - 30Mbps": "LINKS DE INTERNET BANDA LARGA 30MBPS (LIB) COM CPE MODEM",
    "LIB - 50Mbps": "LINKS DE INTERNET BANDA LARGA 50MBPS (LIB) COM CPE MODEM",
    "LIB - 100Mbps": "LINKS DE INTERNET BANDA LARGA 100MBPS (LIB) COM CPE MODEM",
    "LIB - 200Mbps": "LINKS DE INTERNET BANDA LARGA 200MBPS (LIB) COM CPE MODEM",
    "LIB - 300Mbps": "LINKS DE INTERNET BANDA LARGA 300MBPS (LIB) COM CPE MODEM"
}
```

```python
df_final = pd.concat([df_total_equipamento, total_link_simetrico, total_link_banda_larga])
```
